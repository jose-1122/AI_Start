# seaborn 의 내장 데이터인 penguins 정보 확인

```python
import seaborn as sns

print(sns.__version__) # 버전확인
```

```python
import seaborn as sns

penguins = sns.load_dataset("penguins")
penguins
```

```python
print(penguins.info())  # 정보(타입, 행렬, Non-Null Count, dtype, memory usage)
```

```python
print(type(penguins))   # 타입 ex. pandas - DataFrame
print(penguins.shape)   # 행렬
print(penguins.columns) # 컬럼명, dtype
```

```python
penguins.head()         # 앞의 데이터 살펴보기. 괄호 안에 숫자만큼 출력. Default == 5
```

```python
penguins.tail()         # 뒤의 데이터 살펴보기. 괄호 안에 숫자만큼 출력. Default == 5
```

```python
penguins.describe()     # 수치형 데이터 요약(개수, 평균, 표준 편차, 최소값, 최대값)
```

```python
penguins.isnull().sum() # 결측치(missing data) 즉 비어있어 측정되지 못하는 데이터가 몇 개인지 확인.
```

# csv, excel 파일로 만들기

```python
import seaborn as sns

penguins = sns.load_dataset("penguins")
```

```python
penguins.to_csv("penguins.csv", index=False)      # csv 파일로 만들기
penguins.to_excel("penguins.xlsx", index=False)   # excel 파일로 만들기
```
