```python
###   Node 클래스   ###
class Node:
    def __init__(self, data, next = None):
        self.data = data
        self.next = next
        
    def __repr__(self):
        return "({})".format (self.data)           # return "(%s)" % self.data 와 같음. The queue : (a) 로 출력.
        #return "({})".format(repr(self.data))      # return "(%s)" % repr(self.data) 와 같음. The queue : ('a') 로 출력.



###   QueueUsingSinglyLinkedList (단일 연결 리스트 기반 큐) 클래스   ###
class QueueUsingSinglyLinkedList:
    def __init__(self):
        self.head = None                           # head node 존재.
        self.node_counter = 0

    def enqueue(self, data):                       # enqueue(data) method 존재. 즉 큐의 앞쪽(head)에 노드 추가.
        print("\n노드 %s 추가(enqueue)" %str(data))
        if self.head == None:
            self.head = Node(data)
        else:
            new_node = Node(data)
            new_node.next = self.head
            self.head = new_node
            #del new_node
        self.node_counter += 1

    def dequeue(self):                             # dequeue() method 존재. 즉 큐의 뒤쪽(tail)에 노드 반환 및 삭제.
        print("가장 뒤 노드 삭제(dequeue)")
        if self.head == None:
            print("The queue is empty, so cannot dequeue.")
        else:
            prev_node = self.head
            cursor = prev_node.next
            if cursor == None:
                self.head = None
            else:
                while cursor.next != None:
                    prev_node = cursor
                    cursor = cursor.next
                prev_node.next = cursor.next
            #del prev_node
            #del cursor
            self.node_counter -= 1

    def peek(self):                                # peek() method 존재. 즉 큐의 가장 앞쪽(head) 노드 확인. 삭제 X.
        if self.head != None:
            return self.head.data
        else:
            return ""
    
    def print_list(self):                          # print_list() method 존재. 방법 1). 즉 stack 의 가장 위에 node 부터 순차적으로 출력.
        print("The queue : ", end = "")
        cursor = self.head
        if self.head == None:
            print("")
        else:
            print("%s" % self.head, end = "")
            while cursor.next != None:
                print(" - %s" % cursor.next, end = "")
                cursor = cursor.next
        #del cursor
    
    def length(self):                                     # length() method 존재. 즉 queue 에 있는 node 가 총 몇 개인지를 반환.
        return self.node_counter
    
    def isEmpty(self):                                    # isEmpty() method 존재. 즉 queue 가 비어있으면 True 를, 비어있지 않으면 False 를 반환.
        if self.head == None:
            return True
        else: 
            return False

if __name__ == "__main__":
    Queue = QueueUsingSinglyLinkedList()
    
    print("The queue's length :", Queue.length(), end = "\n")
    Queue.print_list()
    
    Queue.enqueue(1)
    print("The queue's length :", Queue.length(), end = "\n")
    Queue.print_list()
    print("\nThe queue's top element is", Queue.peek())
    
    Queue.enqueue(4)
    print("The queue's length :", Queue.length(), end = "\n")
    Queue.print_list()
    print("\nThe queue's top element is", Queue.peek())
    
    Queue.enqueue("a")
    print("The queue's length :", Queue.length(), end = "\n")
    Queue.print_list()
    print("\nThe queue's top element is", Queue.peek())
    
    Queue.enqueue("^")
    print("The queue's length :", Queue.length(), end = "\n")
    Queue.print_list()
    print("\nThe queue's top element is", Queue.peek())
    
    print("\n\nThe queue is empty ?", Queue.isEmpty())
    Queue.dequeue()
    print("The queue's length :", Queue.length(), end = "\n")
    Queue.print_list()
    print("\nThe queue's top element is", Queue.peek())
    
    print("\n\nThe queue is empty ?", Queue.isEmpty())
    Queue.dequeue()
    print("The queue's length :", Queue.length(), end = "\n")
    Queue.print_list()
    print("\nThe queue's top element is", Queue.peek())
    
    print("\n\nThe queue is empty ?", Queue.isEmpty())
    Queue.dequeue()
    print("The queue's length :", Queue.length(), end = "\n")
    Queue.print_list()
    print("\nThe queue's top element is", Queue.peek())
    
    print("\n\nThe queue is empty ?", Queue.isEmpty())
    Queue.dequeue()
    print("The queue's length :", Queue.length(), end = "\n")
    Queue.print_list()
    print("The queue's top element is", Queue.peek())
    
    print("\n\nThe queue is empty ?", Queue.isEmpty())
    Queue.dequeue()
```
