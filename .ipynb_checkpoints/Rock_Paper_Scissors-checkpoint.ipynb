```python
# 가위바위보 게임. 3판 2선승제. (컴퓨터 vs 나)
# 0 : 컴퓨터 승, 1 : 컴퓨터 패, 2 : 무승부.
# 파이썬 연산자 우선순위 중요. 왼쪽으로 갈수록 우선순위가 높다.
# &(비트연산자 AND) > |(비트연산자 OR) > in, not in, is, is not, <, <=, >, >=, !=, ==(비교연산자) > not x, and, or(논리연산자)

def RockPaperScissors(com, me):
    a = ["가위", "바위", "보"]
    print("나 : " + a[me], "vs", "컴퓨터 : " + a[com])
    
    if (me==2 and com==0) or (me==0 and com==1) or (me==1 and com==2):
        print("졌습니다.")
        #print("me : %d, com : %d, choose : %d" % (me, com, choose))              # 참고.
        return 0
    elif (me==0 and com==2) or (me==1 and com==0) or (me==2 and com==1):
        print("이겼습니다.")
        #print("me : %d, com : %d, choose : %d" % (me, com, choose))              # 참고.
        return 1
    else: #(me==0 and com==0) or (me==1 and com==1) or (me==2 and com==2)
        print("비겼습니다.")
        #print("me : %d, com : %d, choose : %d" % (me, com, choose))              # 참고.
        return 2

import random

com_score = 0
me_score = 0

print("[ 3판 2선승제 가위바위보 게임을 시작합니다. ]\n\n")

while (me_score != 2) and (com_score != 2):
    choose = int(input("가위(0), 바위(1), 보(2) 중 하나를 선택해주세요.(그만하려면 9를 입력하세요.) "))
    com = random.randrange(0, 3)
    
    if (choose == 0) or (choose == 1) or (choose == 2):
        count = RockPaperScissors(com, choose)
        #print("com : %d, choose : %d" % (com, choose))                           # 참고.
        if count == 0:
            com_score += 1
            print("현재 %d대 %d\n\n" % (me_score, com_score))
        elif count == 1:
            me_score += 1
            print("현재 %d대 %d\n\n" % (me_score, com_score))
        else:
            print("현재 %d대 %d\n\n" % (me_score, com_score))
    elif choose == 9:
        break
    else:
        print("다시 선택해 주세요.(가위(0), 바위(1), 보(2), 종료(9)) \n")
        continue

if com_score == 2:
  print("컴퓨터 승! 게임 끝!!!")
elif me_score == 2:
  print("컴퓨터 패! 게임 끝!!!")
```
