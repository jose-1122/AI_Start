```python
###   Node 클래스   ###
class Node:
    def __init__(self, data, next = None):
        self.data = data
        self.next = next

    def __repr__(self):
        return "({})".format (self.data)           # return "(%s)" % self.data 와 같음. The stack : (a) 로 출력.
        #return "({})".format(repr(self.data))      # return "(%s)" % repr(self.data) 와 같음. The stack : ('a') 로 출력.
        
        
        
###   StackUsingSinglyLinkedList (단일 연결 리스트 기반 스택) 클래스   ###
class StackUsingSinglyLinkedList:
    def __init__(self):
        self.head = None                           # head node 존재.
        self.node_counter = 0

    def push(self, data):                          # push(data) method 존재. 즉 stack 의 가장 위에 node 추가.
        print("\n노드 %s 추가(push)" %str(data))
        if self.head == None:
            self.head = Node(data)
        else:
            new_node = Node(data)
            new_node.next = self.head
            self.head = new_node
            #del new_node
        self.node_counter += 1
        
    def pop(self):                                 # pop() method 존재. 즉 stack 의 가장 위에 node 삭제.
        print("가장 위 노드 삭제(pop)")
        if self.head != None:
            self.head = self.head.next
            self.node_counter -= 1
        else:
            print("The stack is empty, so cannot pop.")

    def top(self):                                 # top() method 존재. 즉 stack 의 가장 위에 node 반환. 삭제 X.
        if self.head != None: 
            return  self.head.data
        else:
            return ""
    
    def print_list(self):                          # print_list() method 존재. 방법 1). 즉 stack 의 가장 위에 node 부터 순차적으로 출력.
        print("The stack : ", end = "")
        cursor = self.head
        if self.head == None:
            print("")
        else:
            print("%s" % self.head, end = "")
            while cursor.next != None:
                print(" - %s" % cursor.next, end = "")
                cursor = cursor.next
        #del cursor
    '''
    def print_list(self):                          # print_list() method 존재. 방법 2).
        output = ""
        if self.head != None:
            output = "%s" % self.head
            cursor = self.head
            while cursor.next != None:
                output += " -> %s" % cursor.next
                cursor = cursor.next
        print(output)
    '''
    def length(self):                              # length() method 존재. 즉 stack 에 있는 node 가 총 몇 개인지를 반환.
        return self.node_counter
    
    def isEmpty(self):                             # isEmpty() method 존재. 즉 stack 이 비어있으면 True 를, 비어있지 않으면 False 를 반환.
        if self.head == None: 
            return True
        else: 
            return False
            
if __name__ == "__main__":
    Stack = StackUsingSinglyLinkedList()
    
    print("The stack's length :", Stack.length(), end = "\n")
    Stack.print_list()

    Stack.push(1)
    print("The stack's length :", Stack.length(), end = "\n")
    Stack.print_list()
    print("\nThe stack's top element is", Stack.top())

    Stack.push("a")
    print("The stack's length :", Stack.length(), end = "\n")
    Stack.print_list()
    print("\nThe stack's top element is", Stack.top())

    Stack.push(10)
    print("The stack's length :", Stack.length(), end = "\n")
    Stack.print_list()
    print("\nThe stack's top element is", Stack.top())
    
    Stack.push("@")
    print("The stack's length :", Stack.length(), end = "\n")
    Stack.print_list()
    print("\nThe stack's top element is", Stack.top())

    print("\n\nThe stack is empty ?", Stack.isEmpty())
    print("The stack's top element is", Stack.top())
    Stack.pop()
    print("The stack's length :", Stack.length(), end = "\n")
    Stack.print_list()
    print("\nThe stack's top element is", Stack.top())

    print("\n\nThe stack is empty ?", Stack.isEmpty())
    print("The stack's top element is", Stack.top())
    Stack.pop()
    print("The stack's length :", Stack.length(), end = "\n")
    Stack.print_list()
    print("\nThe stack's top element is", Stack.top())
    
    print("\n\nThe stack is empty ?", Stack.isEmpty())
    print("The stack's top element is", Stack.top())
    Stack.pop()
    print("The stack's length :", Stack.length(), end = "\n")
    Stack.print_list()
    print("\nThe stack's top element is", Stack.top())
    
    print("\n\nThe stack is empty ?", Stack.isEmpty())
    print("The stack's top element is", Stack.top())
    Stack.pop()
    print("The stack's length :", Stack.length(), end = "\n")
    Stack.print_list()
    print("The stack's top element is", Stack.top())
  
    print("\n\nThe stack is empty ?", Stack.isEmpty())
    Stack.pop()
```
