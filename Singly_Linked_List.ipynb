```python
###   Node 클래스   ###
class Node:
    def __init__(self, data, next = None):
        # Assumption : None data shouldn't be allowed for security reasons.
        message = "Inserting %s value data is not allowed." % data
        assert data != None, message               # data 가 None 이 아니라면 아래 명령문을 실행. data 가 None 이라면 message 를 출력한 뒤 종료.
        
        self.data = data
        self.next = next
        
    def __repr__(self):
        return "({})".format (self.data)            # return "(%s)" % self.data 와 같음. The stack : (a) 로 출력.
        #return "({})".format(repr(self.data))      # return "(%s)" % repr(self.data) 와 같음. The stack : ('a') 로 출력.



###   Singly Linked List(단일 연결 리스트) 클래스   ###
class SinglyLinkedList:
    def __init__(self):                             # 생성자.
        self.head = None                            # head node 존재.
        self.node_counter = 0
    
    def add_node(self, data):                       # add_node(data) method 존재. 방법 1). 즉 head 에 node 추가. 가장 기본적이고 단일연결리스트의 기본 개념에 가장 부합하는 방법.
        print("\n노드 %s 추가" %str(data))
        if self.head == None:
            self.head = Node(data)
        else:
            self.head = Node(data, self.head)
        self.node_counter += 1
    '''
    def add_node(self, data):                        # add_node(data) method 존재. 방법 2). 즉 next 가 None 인 node (= 위치적으로 linked list 의 tail 에, 즉 가장 뒤에 있음)를 찾아 data 를 값으로 하는 새 node 를 추가. 매번 tail 을 찾는 과정을 거쳐야 하기 때문에 효율적이지 않음.
        print("\n노드 %s 추가" %str(data))
        cursor = self.head
        if self.head == None:
            self.head = Node(data)
        else:
            while cursor.next != None:
              cursor = cursor.next
            cursor.next = Node(data)
        #del cursor
        self.node_counter += 1
    '''
    def delete_node(self, data):                    # delete_node(data) method 존재. 방법 1). 즉 node 의 인덱스를 가지고 삭제를 하는 것이 아니라, 삭제하고자 하는 노드를 head 에서부터 검색을 하다가 삭제하고자 하는 node 를 찾았을때 그 node 를 제거. (즉 head 의 노드를 삭제하는 delete_head 와 next 가 None 인 node (= 위치적으로 linked list 의 tail 에, 즉 가장 뒤에 있음)를 삭제하는 delete_tail 은 불필요함.)
        print("\n노드 %s 삭제" %str(data))
        cursor = self.head
        if self.head.data == data:
            self.head = self.head.next
        else:
            while cursor.next.data != data:
                cursor = cursor.next
            cursor.next = cursor.next.next
        #del cursor
        self.node_counter -= 1
    '''
    def delete_node(self, data):                     # delete_node(data) method 존재. 방법 2).
        print("\n노드 %s 삭제" %str(data))
        cursor = self.head
        if self.head.data == data:
            if self.node_counter > 1:
                self.head = cursor.next
            else:
                self.head = None
            self.node_counter -= 1
        else:
            prev_node = self.head
            cursor = prev_node.next
            while cursor != None:
                if cursor.data == data:
                    prev_node.next = cursor.next
                    self.node_counter -= 1
                    break
                else:
                    prev_node = cursor
                    cursor = cursor.next
        #del prev_node
        #del cursor
    '''    
    def is_in(self, data):                          # is_in(data) method 존재. 즉 head 에서부터 data 와 같은 값을 가지는 node 가 있는지 순차적으로 검색. 있으면 True, 없으면 False 를 반환. Membership query 를 위함.
        cursor = self.head
        while cursor != None:
            if cursor.data == data:
                #del cursor
                return True
            else:
                cursor = cursor.next
        #del cursor
        return False
   
    def print_list(self):                           # print_list() method 존재. 즉 head 에서부터 모든 node 를 순차적으로 출력.
        print("The Singly Linked List : ", end = "")
        cursor = self.head
        if self.head == None:
            print("")
        else:
            print("%s" % self.head, end = "")
            while cursor.next != None:
                print(" - %s" % cursor.next, end = "")
                cursor = cursor.next
            print()
        #del cursor 
    '''
    def print_list(self):                            # print_list() method 존재. 방법 2).
        output = ""
        if self.head != None:
            output = "%s" % self.head
            cursor = self.head
            while cursor.next != None:
                output += " -> %s" % cursor.next
                cursor = cursor.next
        print(output)
    '''
    def length(self):                               # length() method 존재. 즉 linked list 에 있는 node 가 총 몇 개인지를 반환.
        return self.node_counter

if __name__ == "__main__":
    SLL = SinglyLinkedList()

    print("The SLL's length :" , SLL.length())
    SLL.print_list()
    
    SLL.add_node("k")
    print("The SLL's length :" , SLL.length())
    SLL.print_list()
    
    SLL.add_node(4)
    print("The SLL's length :" , SLL.length())
    SLL.print_list()
    
    SLL.add_node("@")
    print("The SLL's length :" , SLL.length())
    SLL.print_list()

    SLL.add_node(7)
    print("The SLL's length :" , SLL.length())
    SLL.print_list()
    
    SLL.add_node("E")
    print("The SLL's length :" , SLL.length())
    SLL.print_list()
    
    print()
    print("3 is in? {}\n".format(SLL.is_in(3)))
    print("k is in? {}\n".format(SLL.is_in("k")))
    print("a is in? {}\n".format(SLL.is_in("a")))
    print("E is in? {}\n".format(SLL.is_in("E")))
    print("@ is in? {}  ".format(SLL.is_in("@")))

    SLL.delete_node("@")
    print("The SLL's length :" , SLL.length())
    SLL.print_list()
    
    SLL.delete_node("E")
    print("The SLL's length :" , SLL.length())
    SLL.print_list()
    
    SLL.delete_node(4)
    print("The SLL's length :" , SLL.length())
    SLL.print_list()

    SLL.delete_node("k")
    print("The SLL's length :" , SLL.length())
    SLL.print_list()

    SLL.delete_node(7)
    print("The SLL's length :" , SLL.length())
    SLL.print_list()
    
    #SLL.add_node(None)                             # None 을 add_node(data) 시 Assumption 에러가 발생.
```
