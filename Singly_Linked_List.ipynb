```python
###   Node 클래스   ###
class Node:
    def __init__(self, data = None, next = None):
        self.data = data
        self.next = next



###   Singly Linked List(단일 연결 리스트) 클래스   ###
class SinglyLinkedList:
    def __init__(self):                                         # 생성자.
        self.head = None                                        # head node 존재.
        self.count = 0
    
    def add_node(self, data):                                   # add_node(data) method 존재. 즉 head 에 노드 추가.
        if self.head == None:
            self.head = Node(data)
        else:
            self.head = Node(data, self.head)
        self.count += 1
        print("{0:^17}".format("[ 노드" + str(data) + " 추가 ]"))

    def delete_node(self, data):                                # delete_node(data) method 존재. 즉 head 부터 검색을 하다가 원하는 node 를 찾았을때 그 node 를 제거.
        index = self.head
        if self.head.data == data:
           self.head = self.head.next
           index = None
        else:
          while index.next.data != data:
            index = index.next
          index.next = index.next.next
        self.count -= 1
        print("{0:^17}".format("[ 노드" + str(data) + " 삭제 ]"))

    def is_in(self, data):                                      # is_in(data) method 존재. True 혹은 False 를 반환. Membership query 를 위함.
        index = self.head
        while index != None:
            if (index.data == data):
                return True
            index = index.next
        return False

    def print_list(self):                                       # print_list() method 존재. 즉 head 에서 부터 모든 node 를 순차적으로 출력.
        index = self.head
        print("The Singly Linked List : ", end = "")
        if index == None:
          print("None")
        else:
          while index:
              if index.next != None:
                  print(index.data, "-> ", end = "")
              else:
                  print(index.data)
              index = index.next

    def length(self):                                           # length() method 존재. 즉 모든 node 의 수를 반환.
        return self.count

if __name__ == "__main__":
    SLL = SinglyLinkedList()

    SLL.print_list()
    print("The SLL's length :" , SLL.length(), "\n")

    SLL.add_node(1)
    SLL.print_list()
    print("The SLL's length :" , SLL.length(), "\n")
    
    SLL.add_node("A")
    SLL.print_list()
    print("The SLL's length :" , SLL.length(), "\n")
    
    SLL.add_node(3)
    SLL.print_list()
    print("The SLL's length :" , SLL.length(), "\n")

    SLL.add_node("w")
    SLL.print_list()
    print("The SLL's length :" , SLL.length(), "\n")
    
    SLL.add_node(5)
    SLL.print_list()
    print("The SLL's length :" , SLL.length(), "\n")
    
    #print("    3 is in? {} \n" .format(SLL.is_in(3)))
    print("    3 is in? ", SLL.is_in(3), "\n")
    print("    7 is in? ", SLL.is_in(7), "\n")
    print("    1 is in? ", SLL.is_in(1), "\n")
    print("    w is in? ", SLL.is_in("w"), "\n")
    print("    0 is in? ", SLL.is_in(0), "\n")

    SLL.delete_node(5)
    SLL.print_list()
    print("The SLL's length :" , SLL.length(), "\n")

    SLL.delete_node(1)
    SLL.print_list()
    print("The SLL's length :" , SLL.length(), "\n")

    SLL.delete_node(3)
    SLL.print_list()
    print("The SLL's length :" , SLL.length(), "\n")

    SLL.delete_node("A")
    SLL.print_list()
    print("The SLL's length :" , SLL.length(), "\n")

    SLL.delete_node("w")
    SLL.print_list()
    print("The SLL's length :" , SLL.length(), "\n")
```
