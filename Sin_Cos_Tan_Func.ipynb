```python
import matplotlib.pyplot as plt
import numpy as np
import math

# np.linspace(시작값, 끝값, 개수)
x = np.linspace(0, 2 * math.pi, 100)
y = np.sin(x)

plt.plot(x, y, 'o')
plt.title("sin graph")
plt.xlabel("[0, 2pi]")
plt.ylabel("y = sin(x)")
```

```python
x = np.linspace(0, 2 * math.pi, 50)
y = np.cos(x)

plt.plot(x, y, 'o')
plt.title("cos graph")
plt.xlabel("[0, 2pi]")
plt.ylabel("y = cos(x)")
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000) # np.pi == math.pi
y = np.tan(x)

plt.plot(x, y, 'o')
plt.ylim(-10, 10)
plt.title("tan graph")
plt.xlabel("[-2pi, 2pi]")
plt.ylabel("y = tan(x)")
```

```python
x = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x)
y2 = 2 * np.sin(x)
y3 = 4 * np.sin(x)

plt.figure(figsize=(6,4))             # 그래프 크기 설정
plt.plot(x, y1, 'o', label="sin(x)")  # 그래프에 범례 표시 
plt.plot(x, y2, '^', label="2sin(x)") 
plt.plot(x, y3,      label="4sin(x)") 
plt.legend()                          # label 표시 유무 설정
plt.title("Three sin graphs")
plt.xlabel("[0, 4pi]")
plt.ylabel("y")
```
