```python
from tkinter import * # tkinter : GUI 에 대한 표준 Python 인터페이스
import random


# 메인
window=Tk()                           # [] = Tk() : 가장 상위 레벨의 윈도우 창을 생성.
window.title("Path planning for UAV to its charging stations") # 윈도우이름.title("제목") : 윈도우 창의 제목.
window.geometry("1000x700+330+150")   # [].geometry("창 너비 x 창 높이, (+/-) 창 가로방향 위치, (+/-) 창 세로방향 위치").
window.resizable(True, False)         # [].resizeable(상하, 좌우) : 윈도우 창의 크기 조절. True로 설정할 경우 조절할 수 있다.


# 캔버스
Area_colour="skyblue"
Area_width=950
Area_height=600
Area = Canvas(bg=Area_colour, width=Area_width, height=Area_height) # 캔버스 생성. 캔버스 제일 위 왼쪽 좌표 == (0,0)
Area.pack()                                                         # 캔버스 배치


# 위젯
label = Label(window, text="[ Let's head over to the destination! Our goal is for UAV to reach the destination as few tries as possible. ]", height=2, fg="black")
label.pack()   # 위젯 배치


# 메시지
message=tkinter.Message(window, text="[ How to control : Press arrow keys. ]     [ Black square : UAV / Red X : Destination / Green stuffs : Obstacle]", width=1000)
message.pack() # 메시지 배치


# 버튼
button = Button(window, text="Reset", width=30, height=2, overrelief="solid", bg="white", fg="green", command=reset)
button.pack()  # 버튼 배치


# 목표지점 및 장애물
for i in range(100):
    UAV_x = random.randint(25, 975)
    UAV_y = random.randint(0, 500)
    
    if i==0:
        UAV_y = random.randint(0, 100)
        # [].create_line(시작x위치, 시작 y위치, 끝x위치, 끝y위치, 선의 굵기, 선의 색깔)
        Area.create_line(UAV_x, UAV_y, UAV_x+12, UAV_y+16, width=5, fill="red")
        UAV_y += 16
        Area.create_line(UAV_x, UAV_y, UAV_x+12, UAV_y-16, width=5, fill="red")
    elif i<=20:
        Area.create_line(UAV_x, UAV_y,  UAV_x+random.randrange(1, 100), UAV_y+random.randrange(1, 100), width=10, fill="green")
    elif i<=40:
        Area.create_line(UAV_x, UAV_y,  UAV_x+random.randrange(1, 100), UAV_y-random.randrange(1, 100), width=10, fill="green")
    elif i<=60:
        Area.create_polygon(UAV_x, UAV_y,  UAV_x+random.randrange(1, 100), UAV_y+random.randrange(1, 100), width=10, fill="green")
    elif i<=80:
        Area.create_oval(UAV_x, UAV_y,  UAV_x+random.randrange(1, 100), UAV_y+random.randrange(1, 100), width=1, fill="green")
    else:
        Area.create_arc(UAV_x, UAV_y,  UAV_x+random.randrange(1, 100), UAV_y+random.randrange(1, 100), width=1, fill="green")

        
# 초기화
UAV_x = Area_width/2
UAV_y = Area_height
count=0

Area.create_line(UAV_x, UAV_y,  UAV_x, UAV_y-10, width=10, fill="black")
UAV_y -= 10


def reset():
    Area.delete(ALL)    # 캔버스 초기화
    global UAV_x
    global UAV_y
    global count
    for i in range(100):
        UAV_x = random.randint(25, 975)
        UAV_y = random.randint(0, 500)
        
        if i==0:
            UAV_y = random.randint(0, 100)
            Area.create_line(UAV_x, UAV_y, UAV_x+12, UAV_y+16, width=5, fill="red")
            UAV_y += 16
            Area.create_line(UAV_x, UAV_y, UAV_x+12, UAV_y-16, width=5, fill="red")
        elif i<=20:
            Area.create_line(UAV_x, UAV_y,  UAV_x+random.randrange(1, 100), UAV_y+random.randrange(1, 100), width=10, fill="green")
        elif i<=40:
            Area.create_line(UAV_x, UAV_y,  UAV_x+random.randrange(1, 100), UAV_y-random.randrange(1, 100), width=10, fill="green")
        elif i<=60:
            Area.create_polygon(UAV_x, UAV_y,  UAV_x+random.randrange(1, 100), UAV_y+random.randrange(1, 100), width=10, fill="green")
        elif i<=80:
            Area.create_oval(UAV_x, UAV_y,  UAV_x+random.randrange(1, 100), UAV_y+random.randrange(1, 100), width=1, fill="green")
        else:
            Area.create_arc(UAV_x, UAV_y,  UAV_x+random.randrange(1, 100), UAV_y+random.randrange(1, 100), width=1, fill="green")

    UAV_x = Area_width/2
    UAV_y = Area_height
    count = 0
    
    Area.create_line(UAV_x, UAV_y,  UAV_x, UAV_y-10, width=10, fill="black")
    label.config(text=" Wanna give it another try? Press arrow keys to play again. ")

    
def countUp():
    global count
    count += 1
    if count%10==1:
        label.config(text=str(count) + "st try")
    elif count%10==2:
        label.config(text=str(count) + "nd try")
    elif count%10==3:
        label.config(text=str(count) + "rd try")
    else:
        label.config(text=str(count) + "th try")

        
def UAV_move_forward(event):
    global UAV_y
    Area.create_line(UAV_x, UAV_y,  UAV_x, UAV_y-10, width=10, fill="white")
    UAV_y -= 10
    Area.create_line(UAV_x, UAV_y,  UAV_x, UAV_y-10, width=10, fill="black")
    countUp()
    
def UAV_move_backward(event):
    global UAV_y
    Area.create_line(UAV_x, UAV_y,  UAV_x, UAV_y+10, width=10, fill="white")
    UAV_y += 10
    Area.create_line(UAV_x, UAV_y,  UAV_x, UAV_y+10, width=10, fill="black")
    countUp()

def UAV_move_left(event):
    global UAV_x
    Area.create_line(UAV_x, UAV_y,  UAV_x-10, UAV_y, width=10, fill="white")
    UAV_x -= 10
    Area.create_line(UAV_x, UAV_y,  UAV_x-10, UAV_y, width=10, fill="black")
    countUp()

def UAV_move_right(event):
    global UAV_x
    Area.create_line(UAV_x, UAV_y,  UAV_x+10, UAV_y, width=10, fill="white")
    UAV_x += 10
    Area.create_line(UAV_x, UAV_y,  UAV_x+10, UAV_y, width=10, fill="black")
    countUp()
    
    
# 바인드
window.bind("<Up>", UAV_move_forward)   # [].bind(키, 함수) : 어떤 키가 눌리면 함수를 실행.
window.bind("<Down>", UAV_move_backward)
window.bind("<Left>", UAV_move_left)
window.bind("<Right>", UAV_move_right)


window.mainloop()     # [].mainloop() : 윈도우 창이 윈도우가 종료될 때까지 실행.
```
