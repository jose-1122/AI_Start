```python
import matplotlib.pyplot as plt
import numpy as np

# np.linspace(시작값, 끝값, 개수)
x = np.linspace(0, 2*np.pi, 50) # np.pi == math.pi
y = np.sin(x)

plt.plot(x, y, ">")
plt.title("sin graph")
plt.xlabel("x = [0, 2pi]")
plt.ylabel("y = sin(x)")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 50)
y = np.cos(x)

plt.plot(x, y, 'x')
plt.title("cos graph")
plt.xlabel("x = [0, 2pi]")
plt.ylabel("y = cos(x)")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2*np.pi, 2*np.pi, 500)
y = np.tan(x)

plt.plot(x, y, "+")
plt.ylim(-10, 10)
plt.title("tan graph")
plt.xlabel("x = [-2pi, 2pi]")
plt.ylabel("y = tan(x)")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1*np.pi, 3*np.pi, 100)
y1 = 5*np.sin(x)
y2 = 3*np.cos(x)
y3 = -1*np.tan(x)

plt.figure(figsize=(12,4))             # 그래프 크기 설정. (Default == 6, 4)
plt.plot(x, y1, "*", label="5sin(x)")  # 그래프에 범례 표시 
plt.plot(x, y2, ".", label="3cos(x)") 
plt.plot(x, y3, "v", label="-tan(x)")
plt.ylim(-5, 5)
plt.legend()                           # label 표시 유무 설정

plt.title("Three graphs")
plt.xlabel("x = [-pi, 3pi]")
plt.ylabel("y")
```
