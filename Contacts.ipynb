```python
class Contact:
    def __init__(self, name, phone_number, e_mail, addr):
        self.name = name
        self.phone_number = phone_number
        self.e_mail = e_mail
        self.addr = addr
        
    def print_info(self):
        print("\nName : ", self.name)
        print("Phone_number : ", self.phone_number)
        print("E-maile : ", self.e_mail)
        print("Address : ", self.addr)
        print("\n{0:-^15}".format(""))

class Work:
    # 메뉴
    def print_menu(self):
        while True:
            print("\n1. 연락처 입력")
            print("2. 연락처 출력")
            print("3. 연락처 삭제")
            print("4. 연락처 조회")
            print("5. 종료\n")
            menu = int(input("메뉴를 선택하세요(1~5) : "))
            if not ((menu>=1) and (menu<=5)):
                print("\n1, 2, 3, 4, 5 중 선택해주세요.")
                print("\n{0:=^15}".format(""))
                continue
            else:
                return menu
    
    # 연락처 입력
    def set_contact(self):
        name = input("이름 : ")
        phone_num = input("휴대폰 번호 : ")
        e_mail = input("이메일 : ")
        addr = input("주소 : ")
        contact = Contact(name, phone_num, e_mail, addr)
        print("\n{0:=^15}".format(""))
        return contact
    
    # 연락처 출력
    def print_contact(self, contact_list):
        if contact_list==[]:
            print("\n저장된 연락처가 없습니다.")
            print("\n{0:-^15}".format(""))
        else:
            for contact in contact_list:
                contact.print_info()
    
    # 연락처 삭제
    def delete_contact(self, contact_list, name):
        if contact_list==[]:
            print("연락처가 비어있어 삭제할 연락처가 없습니다.")
        else:
            for i, contact in enumerate(contact_list):
                if contact.name == name:
                    del contact_list[i]
                    print(name + "님이 연락처에서 삭제되었습니다.")
                    return 0
            else:
                print(name + "님은 연락처에 없습니다.")
    
    # 연락처 조회
    def is_in(self, contact_list, name):
        if contact_list==[]:
            print("연락처가 비어있어", name + "님은 연락처에 없습니다.")
        else:
            for i, contact in enumerate(contact_list):
                if contact.name == name:
                    print(name + "님은 연락처에 있습니다.")
                    return 0
            else:
                print(name + "님은 연락처에 없습니다.")
                
if __name__=="__main__":
    w = Work()
    
    print("{0:=^15}".format(""))
    print("{0:^12}".format("[  주소록  ]"))
    print("{0:=^15}".format(""))
    contact_list = []

    while True:
        sel_num = w.print_menu()
        if sel_num==1:
            print("저장할 연락처를 입력하세요.")
            c = w.set_contact()
            contact_list.append(c)
        elif sel_num==2:
            print("다음은 저장된 연락처입니다.")
            print("\n{0:-^15}".format(""))
            w.print_contact(contact_list)
        elif sel_num==3:
            print("연락처를 삭제하겠습니다.")
            del_name = input("삭제할 이름은 : ")
            w.delete_contact(contact_list, del_name)
            print("\n{0:=^15}".format(""))
        elif sel_num==4:
            print("연락처가 있는지 조회합니다.")
            name_is_in = input("조회할 이름은 : ")
            w.is_in(contact_list, name_is_in)
            print("\n{0:=^15}".format(""))
        elif sel_num==5:
            print("종료합니다.")
            break
```
